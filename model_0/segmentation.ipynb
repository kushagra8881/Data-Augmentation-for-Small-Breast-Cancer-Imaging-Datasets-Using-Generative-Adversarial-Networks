```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import os
from scipy.signal import hilbert

# Load the OASBUD.mat file
data = scipy.io.loadmat('OASBUD.mat')
data_content = data['data']

# Constants
c = 1540  # Speed of sound in m/s
fs = 40e6  # Sampling frequency in Hz
width = 50  # Width of the image in mm
eps = 1e-10  # Small constant to prevent log issues

# Create required directories
base_dir = "output_images"
for label in ["0", "1"]:
    os.makedirs(f"{base_dir}/{label}/rf", exist_ok=True)
    os.makedirs(f"{base_dir}/{label}/roi", exist_ok=True)

def compute_envelope(signal):
    """Compute envelope and convert to dB scale"""
    analytic_signal = hilbert(signal, axis=0)
    envelope = np.abs(analytic_signal)
    return 20 * np.log10(envelope + eps)

def save_clean_image(image, path):
    """Save RF image with consistent intensity range"""
    plt.figure(figsize=(8, 6))
    plt.imshow(image, cmap='gray', aspect='auto')
    plt.axis('off')
    plt.clim(40, 80)  # Set consistent intensity range for RF images
    plt.savefig(path, dpi=300, bbox_inches='tight', pad_inches=0)
    plt.close()

def save_mask_image(mask, path):
    """Save binary mask for segmentation"""
    plt.figure(figsize=(8, 6))
    plt.imshow(mask, cmap='gray', aspect='auto', vmin=0, vmax=1)
    plt.axis('off')
    plt.savefig(path, dpi=300, bbox_inches='tight', pad_inches=0)
    plt.close()

# Process all scans
for idx, scan_info in enumerate(data_content[0]):
    # Extract scan data
    rf1 = scan_info[1]  # RF scan plane 1
    rf2 = scan_info[2]  # RF scan plane 2
    roi1 = scan_info[3].astype(bool)  # ROI mask 1
    roi2 = scan_info[4].astype(bool)  # ROI mask 2
    label = int(scan_info[6])  # Ensure label is integer

    # Compute envelopes for RF images
    env_rf1 = compute_envelope(rf1)
    env_rf2 = compute_envelope(rf2)

    # Determine save path based on label
    label_dir = f"{base_dir}/{label}"

    # Save RF images
    save_clean_image(env_rf1, f"{label_dir}/rf/env_rf1_{idx}.png")
    save_clean_image(env_rf2, f"{label_dir}/rf/env_rf2_{idx}.png")

    # Save ROI masks for segmentation
    save_mask_image(roi1, f"{label_dir}/roi/mask1_{idx}.png")
    save_mask_image(roi2, f"{label_dir}/roi/mask2_{idx}.png")

    print(f"Processed scan {idx}")

print("All scans processed successfully.")
```
